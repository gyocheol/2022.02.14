```python
import pandas as pd
```

```python
df = pd.read_csv("2020년_버스노선별_정류장별_시간대별_승하차_인원_정보.csv", encoding='cp949')
df.head()
```

```python
df.info()
```

```python
df = df[["사용년월", "노선번호", "노선명", "표준버스정류장ID", '역명', '7시하차총승객수', '8시하차총승객수', '9시하차총승객수', 
           '10시하차총승객수', '17시승차총승객수', '18시승차총승객수', '19시승차총승객수', '20시승차총승객수']]
df
```

```python
df =  pd.pivot_table(df, index=['사용년월','노선번호','노선명','표준버스정류장ID','역명'], aggfunc = 'sum')
df
```

```python
df.reset_index(inplace=True)
```
